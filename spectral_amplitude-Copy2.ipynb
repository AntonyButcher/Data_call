```python
for tr in st_inst_flt:
  dataonly = tr.data
  spec = np.fft.rfft(dataonly)
  samp_rate=50
  freq = np.fft.fftfreq(len(spec), d=1./samp_rate)
  plt.plot(abs(freq),abs(spec))
  plt.xlim([-1,25])
  plt.ylim([0,0.01])
  plt.suptitle(tr.id)
  plt.show()
    
```

```python
BHE = st_inst_flt[0]
BHN = st_inst_flt[1]
BHZ = st_inst_flt[2]

data_E=BHE.data
data_N=BHN.data
data_Z=BHZ.data

E_fft = np.fft.rfft(data_E)
N_fft = np.fft.rfft(data_N)
Z_fft = np.fft.rfft(data_Z)

samp_rate=50
freq = np.fft.fftfreq(len(E_fft), d=1./samp_rate)

EZ = E_fft/Z_fft
NZ = N_fft/Z_fft
EN = E_fft/N_fft

plt.plot(abs(freq),EZ)
plt.xlim([-1,25])
plt.ylim([-1000,1000])
plt.suptitle(tr.id)
plt.show()

plt.plot(abs(freq),abs(NZ))
plt.xlim([-1,25])
plt.ylim([0,1000])
plt.suptitle(tr.id)
plt.show()

plt.plot(abs(freq),abs(EN))
plt.xlim([-1,25])
plt.suptitle(tr.id)
plt.show()
```
