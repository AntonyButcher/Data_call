```python
from obspy.fdsn import Client
from obspy import UTCDateTime
import numpy as np 
import matplotlib.pyplot as plt
import scipy

%matplotlib inline

client = Client()
```

Download multiple datasets

```python

t1 = UTCDateTime("2010-02-27T06:45:00.000")
t2 = t1 + 120*60
bulk = [("GB", "HTL", "*", "BHZ", t1, t2), ("GB", "HTL", "*", "BHN", t1, t2), ("GB", "HTL", "*", "BHE", t2, t2)]

st_yk = client.get_waveforms_bulk(bulk)
print st_yk 
```

Downloads individual waveforms from IRIS with the instrument response removed and converted to velocity

```python
t = UTCDateTime("2010-02-27T06:45:00.000")
st = client.get_waveforms("GB", "HTL", "*", "BH*", t, t + 120 * 60, attach_response=True)
print st 
```

Plots the downloaded datasets

```python
st.plot()  
```

```python
st_inst=st.copy() 
st_inst.remove_response(output="VEL") 
st_inst.plot()
```

```python
st_inst_flt = st_inst.copy()
st_inst_flt.detrend('linear')
st_inst_flt.taper(max_percentage=0.05, type='cosine')
```

```python
st_inst_flt.filter('bandpass', freqmin=1.0,freqmax=5, corners=2, zerophase=True)
st_inst_flt.plot()
```
