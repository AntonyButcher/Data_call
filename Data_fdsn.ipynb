```python
from obspy.fdsn import Client
from obspy import UTCDateTime
import numpy as np 
import matplotlib.pyplot as plt
import scipy

%matplotlib inline

client = Client()
```

Download multiple datasets

```python

t1 = UTCDateTime("2010-02-27T00:00:00.000")
t2 = t1 + 120*60
bulk = [("GB", "HTL", "*", "BHZ", t1, t2), ("GB", "HTL", "*", "BHN", t1, t2), ("GB", "HTL", "*", "BHE", t2, t2)]

st_yk = client.get_waveforms_bulk(bulk)
print st_yk 
```

Downloads individual waveforms from IRIS with the instrument response removed and converted to velocity

```python
t = UTCDateTime("2010-02-23T00:00:00.000")
st = client.get_waveforms("GB", "HTL", "*", "BH*", t, t + 120 * 60, attach_response=True)
print st 
```

Plots the downloaded datasets

```python
st.plot()  
```

```python
st_inst=st.copy() 
st_inst.remove_response(output="VEL") 
st_inst.plot()
```

```python
st_inst_flt = st_inst.copy()
st_inst_flt.filter('highpass', freq=0.02, corners=2, zerophase=True)
st_inst_flt.plot()
```

```python

st_inst_flt.detrend('linear')
st_inst_flt.taper(max_percentage=0.01, type='cosine')
st_inst_flt.plot()
```

```python
for tr in st_inst_flt:
    tr.plot(type='dayplot')
```

```python
st_inst_flt.filter('bandpass', freqmin=0.5, freqmax=20, corners=2, zerophase=True)
st_inst_flt.plot()
```

```python
for tr in st_inst_flt: 
    tr.write(tr.id + ".sac", format="SAC") 
```

```python
for tr in st_inst_flt:
  dataonly = tr.data
  spec = np.fft.rfft(dataonly)
  samp_rate=50
  freq = np.fft.fftfreq(len(spec), d=1./samp_rate)
  plt.plot(abs(freq),abs(spec))
  plt.xlim([-1,25])
  plt.ylim([0,0.01])
  plt.suptitle(tr.id)
  plt.show()
    
```
