```python
from obspy.fdsn import Client
from obspy import UTCDateTime
import numpy as np 
import matplotlib.pyplot as plt
import scipy

%matplotlib inline

client = Client()
```

Download multiple datasets

```python

t1 = UTCDateTime("2010-02-27T06:45:00.000")
t2 = t1 + 120*60
bulk = [("GB", "HTL", "*", "BHZ", t1, t2), ("GB", "HTL", "*", "BHN", t1, t2), ("GB", "HTL", "*", "BHE", t2, t2)]

st_yk = client.get_waveforms_bulk(bulk)
print st_yk 
```

Downloads individual waveforms from IRIS with the instrument response removed and converted to velocity

```python
t = UTCDateTime("2010-02-27T06:45:00.000")
st = client.get_waveforms("GB", "HTL", "*", "BH*", t, t + 120 * 60, attach_response=True)
print st 
```

Plots the downloaded datasets

```python
st.plot()  
```

```python
st_flt=st.copy() 
st_flt.remove_response(output="VEL") 
st_flt.plot()
```

```python
st_flt.filter('highpass', freq=2.0, corners=1, zerophase=True)
st_flt.plot()
```

```python
for tr in st_flt: 
    tr.write(tr.id + ".sac", format="SAC") 
```

```python
stp = st_flt.copy()
stp.detrend('linear')
stp.taper(max_percentage=0.5, type='cosine')

print stp
```

```python
for tr in stp:
  dataonly = tr.data
  spec = np.fft.rfft(dataonly)
  samp_rate=50
  freq = np.fft.fftfreq(len(spec), d=1./samp_rate)
#  plt.plot(freq,spec)
  plt.plot(freq,abs(spec))
  plt.xlim([-1,25])
  plt.ylim([0,0.001])
  plt.show()
    
```
