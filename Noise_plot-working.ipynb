```python
from obspy.fdsn import Client
import obspy
from obspy import UTCDateTime

from numpy import *
import numpy as np 
import matplotlib.pyplot as plt
import scipy

%matplotlib inline

client = Client()
```

```python
t = UTCDateTime("2010-02-15T00:00:00.000")
st = client.get_waveforms("GB", "HTL", "*", "BHZ", t, t + 3600*12, attach_response=True)
print st 
```

```python
st_inst=st.copy() 
st_inst.merge()
st_inst.remove_response(output="VEL") 
st_inst.filter('highpass', freq=3, corners=2, zerophase=True)
st_inst.taper(max_percentage=0.01, type='cosine')

print(st_inst)
st_inst.plot() 
```

```python
t1=st_inst[0].stats.starttime
t2=st_inst[0].stats.endtime
samp_rate=st_inst[0].stats.sampling_rate  

fin_len=len(st_inst[0])
print len(st_inst[0])
print samp_rate
```

```python
data=np.array((0))
time=np.array((0))
freq=np.array((0))
print time.shape
print st_inst[0]
```

```python
t=t1

while t < t2:
    cut=st_inst[0].slice(t,t+1200)
    tmid=t+600
    t=t+1200
    length = len(cut)-1
    print length

    time_temp=np.zeros(shape=(length,1))
        
    for a in range(0,length):
        time_temp[a]=[tmid.timestamp]

    time=vstack((time,time_temp))

    
    cut_data=cut.data
    cut_spec=np.fft.fft(cut_data,n=length)
    
    data_temp=np.zeros(shape=(length,1),dtype=np.complex)
        
    for b in range(0,length):
        data_temp[b]=cut_spec[b]
    
    data=vstack((data,data_temp))
       
    freq_temp = np.fft.fftfreq(length, d=1./samp_rate)
    
    freq_temp2=np.zeros(shape=(length,1))
        
    for c in range(0,length):
        freq_temp2[c]=freq_temp[c]   
    
    
    freq=vstack((freq,freq_temp2))
    
    
        
    plt.plot(abs(freq_temp),abs(cut_spec))
    plt.xlim([-1,25])
    plt.ylim([0,0.00001])
    plt.show()

    
print time
print len(time)
print data
print len(data)
print freq
print len(freq)
    
time=np.delete(time,0,0)
data=np.delete(data,0,0)
freq=np.delete(freq,0,0)

print time
print len(time)
print data
print len(data)
print freq
print len(freq)
```
